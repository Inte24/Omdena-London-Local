```python
import snscrape.modules.twitter as snt
import pandas as pd
```

```python
labels=[
    "Username",
    "tweet-content",
    "location",
    "Timestamp",
    "URL"
]
```

```python
def Content(query,limit,start_date,end_date,radius):
   """
   Return the "labels" as a dataframe while you take the inputs of time,query and the radius within london
   """
   #empty list that will store everything
   df=[]
   for i,tweet in enumerate(snt.TwitterSearchScraper(f'{query} since:{start_date} until:{end_date} near:"London" within:{radius}').get_items()):
      
      tdf=[
         tweet.user.username,
         tweet.content,
         tweet.user.location,
         tweet.date,
         tweet.url
          ]
      
      df.append(tdf)
      tdf=[]
      if i==limit-1:
        break
   pdf=pd.DataFrame(df,columns=labels)

   return pdf
```

```python
x= Content("mental health",start_date="2022-08-24",end_date="2022-08-31",limit=100,radius="40km")
```
